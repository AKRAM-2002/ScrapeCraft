#Using BeautifulSoup

```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
url = "https://editorial.rottentomatoes.com/guide/popular-movies/"
```

```python
page = requests.get(url)
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content, "html.parser")
print(soup.prettify())
```

```python
movie_title_divs = soup.find_all('div', class_=re.compile(r'article_movie_title'))
```

```python
for movie_title_div in movie_title_divs:
    title = movie_title_div.find('a').text
    print(title)
```

#Using Regular EXPRESSION

```python
movies_regex = re.findall(r'<div class="article_movie_title".*?>\s*<div><h2><a href=".*?">(.*?)</a>', str(soup))
```
